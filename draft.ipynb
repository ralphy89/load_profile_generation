```python
import pandas as pd
import numpy as np

total_res = 14357
df = pd.read_csv("./Data/total_buildings_by_communes.csv")
df_consommation = pd.read_csv("./Data/load_profile.csv")

near_grid = df[df["Category"] == "near-grid"].copy()

# Compute estimated subscribers
abonne_est = np.round(total_res * (near_grid["Buildings"] / 21706), 0)
near_grid.loc[:, "abonnes_estimes"] = abonne_est
near_grid["abonnes_estimes"] = near_grid["abonnes_estimes"].astype("int64")

# Compute non-connected buildings
near_grid.loc[:, "non_raccordes"] = near_grid["Buildings"] - near_grid["abonnes_estimes"]
near_grid
```
